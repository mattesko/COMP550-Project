```python
all_list_of_files = []

for idx, row in all_data[:5000].iterrows():
    ll = row.related_articles
    
    x_val = "Pants on Fire" + ". "
    for i, xs in enumerate(ll):
        x_val += str(xs) + ". "
    all_list_of_files.append(x_val)
```

```python
high_scores = {}

for x in all_list_of_files:
    x = x.split(".")
    
    claim = x[0]
    data = x[1:]

    vect = TfidfVectorizer(min_df=1, stop_words="english")
    tfidf = vect.fit_transform(x)
    pairwise_similarity = tfidf * tfidf.T
    
    b = pairwise_similarity.toarray()[0]

    arg_list = np.argwhere(b > 0.)
    for idx, arg in enumerate(arg_list[1:]):

        high_scores[b[idx+1]] = x[arg[0]]

    max_score = list(reversed(sorted(list(high_scores.keys()))))[:1]
    for m in max_score:
        if m > 0.1:
            print("CLAIM:", claim)
            print()
            print(m, ":", high_scores[m])
            print()
```
